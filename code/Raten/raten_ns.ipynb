```python
ns = defaultParameters["ns"]
valuesOneHundredTimeSteps, valuesLastTimeStep, ratesvr, ratesvt, ratesvm, ratesvq = [], [], [], [], [], []
ratesvimp, ratesvcat, ratescbur, ratescbut, ratescbum, ratescbuq = [], [], [], [], [], []
ratesomegar, ratesdegmr, ratesomegat, ratesdegmt = [], [], [], []
ratesomegam, ratesdegmm, ratesomegaq, ratesdegmq = [], [], [], []
ratesdilution_si, ratesdilution_a = [], []
ratesdilution_mr, ratesdilution_mt, ratesdilution_mm, ratesdilution_mq = [], [], [], []
ratesdilution_cr, ratesdilution_ct, ratesdilution_cm, ratesdilution_cq = [], [], [], []
ratesdilution_pr, ratesdilution_pt, ratesdilution_pm, ratesdilution_pq = [], [], [], []

for _ in np.linspace(0,5,39):#range(0,4):
    ns = ns - 0.01
    m = WeisseBaseModel({'ns':ns})
    print(ns)

    s = modelbase.Simulator(m)
    #                         ['si',       'a',     'mr',     'mt',   'mm',    'mq',    'cr', 'ct', 'cm', 'cq', 'pr',   'pt',    'pm',     'pq']
    weise_values = np.array([31096.192, 43297.502, 8514.558, 49.333, 49.333, 8811.960, 348.675, 0, 357.898, 0, 0.0414, 471.364, 471.364 , 471.364])
    changedValuesOverTimeArray = s.timeCourse(np.linspace(0,3000,100),weise_values);
    ratesvr.append(s.getRate('vr',r= None)) 
    ratesvt.append(s.getRate('vt',r= None))
    ratesvm.append(s.getRate('vm',r= None)) 
    ratesvq.append(s.getRate('vq',r= None)) 
    ratesvimp.append(s.getRate('vimp',r= None))
    ratesvcat.append(s.getRate('vcat',r= None))
    ratescbur.append(s.getRate('cbur',r= None))
    ratescbut.append(s.getRate('cbut',r= None))
    ratescbum.append(s.getRate('cbum',r= None))
    ratescbuq.append(s.getRate('cbuq',r= None))
    ratesomegar.append(s.getRate('omegar',r= None))
    ratesdegmr.append(s.getRate('degmr',r= None))
    ratesomegat.append(s.getRate('omegat',r= None)) 
    ratesdegmt.append(s.getRate('degmt',r= None)) 
    ratesomegam.append(s.getRate('omegam',r= None)) 
    ratesdegmm.append(s.getRate('degmm',r= None)) 
    ratesomegaq.append(s.getRate('omegaq',r= None)) 
    ratesdegmq.append(s.getRate('degmq',r= None)) 
    ratesdilution_si.append(s.getRate('dilution_si',r= None)) 
    ratesdilution_a.append(s.getRate('dilution_a',r= None)) 
    ratesdilution_mr.append(s.getRate('dilution_mr',r= None)) 
    ratesdilution_mt.append(s.getRate('dilution_mt',r= None)) 
    ratesdilution_mm.append(s.getRate('dilution_mm',r= None)) 
    ratesdilution_mq.append(s.getRate('dilution_mq',r= None)) 
    ratesdilution_cr.append(s.getRate('dilution_cr',r= None)) 
    ratesdilution_ct.append(s.getRate('dilution_ct',r= None)) 
    ratesdilution_cm.append(s.getRate('dilution_cm',r= None)) 
    ratesdilution_cq.append(s.getRate('dilution_cq',r= None)) 
    ratesdilution_pr.append(s.getRate('dilution_pr',r= None)) 
    ratesdilution_pt.append(s.getRate('dilution_pt',r= None)) 
    ratesdilution_pm.append(s.getRate('dilution_pm',r= None)) 
    ratesdilution_pq.append(s.getRate('dilution_pq',r= None)) 

    
    
    valuesOneHundredTimeSteps.append(changedValuesOverTimeArray)
    valuesLastTimeStep.append(changedValuesOverTimeArray[-1])
    #lastratevr.append(ra)
    
    
    
    
    
    
```
